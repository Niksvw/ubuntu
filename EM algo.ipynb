```python
import numpy as np
import pandas as pd
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load your dataset (replace 'your_dataset.csv' with your actual file path)
df = pd.read_csv('Salary_data.csv')

# Display the first few rows of the dataset
df.head()
```

```python
# Select the features (columns) for clustering
features = df.columns

# Convert the features to a NumPy array
X = df[features].values

# Specify the number of clusters (adjust as needed)
n_clusters = 3

# Create a Gaussian Mixture Model
gmm = GaussianMixture(n_components=n_clusters, random_state=42)

# Fit the model to the data
gmm.fit(X)

# Predict the cluster assignments for each data point
labels = gmm.predict(X)

# Add cluster labels to the original dataset
df['Cluster'] = labels
```

```python
# Plot the clusters in 2D (assuming 2 features, adapt as needed)
plt.figure(figsize=(8, 6))
sns.scatterplot(x='YearsExperience', y='Salary', hue='Cluster', data=df, palette='viridis', legend='full')
plt.title('Clusters in 2D Feature Space')
plt.show()
```
